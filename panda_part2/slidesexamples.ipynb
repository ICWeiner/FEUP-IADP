```python
import pandas as pd
```

```python
df1 = pd.read_csv('gym_1.csv',sep=';', index_col ='id', parse_dates=['date'])
df1
```

```python
df2 = pd.read_csv('gym_2.csv',sep=';', index_col ='id', parse_dates=['date'])
df2
```

```python
df_concat_cols_outer = pd.concat([df1,df3],axis=1)
df_concat_cols_outer
```
