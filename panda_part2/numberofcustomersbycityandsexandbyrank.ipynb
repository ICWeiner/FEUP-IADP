```python
import pandas as pd
df = pd.read_csv("gym2_q1.csv",sep=";",index_col="id",parse_dates =["date"])
```

```python
df.groupby(['class','sex']).agg(max_age=('age','max'),min_weight=('weight','min'),mean_height=('height','mean')).round(2)
```
