```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,3,4,5,6])
ypoints1 = np.array([5,10,3,11,12,9])
ypoints2 = np.array([8,5,12,16,13,9])

plt.subplot(1,2,1)
plt.plot(xpoints,ypoints1,'o--g',label = "Person 1")
plt.legend()

plt.subplot(1,2,2)
plt.plot(xpoints,ypoints2,'o--b',label = "Person 2")
plt.legend()

#plt.xlabel("Day")

#plt.ylabel("Walking distance(km)")

font1 = {'family':'serif','color':'red','size':20}

#plt.title("Persons workout",fontdict = font1)





plt.show
```

```python
x_index = np.arange(len(xpoints))
plt.bar(x_index -0.1,ypoints1,width= 0.2, label = "Person 1")
plt.bar(x_index + 0.1,ypoints2,width = 0.2, label = "Person 2")

plt.xticks(ticks=x_index,labels=xpoints)
plt.xlabel('Day')
plt.ylabel("Walking distance (km)")
plt.title("Persons workout",fontdict = font1)
plt.legend()
```

```python
x_index = np.arange(len(xpoints))
plt.barh(x_index -0.1,ypoints1,height= 0.2, label = "Person 1")
plt.barh(x_index + 0.1,ypoints2,height = 0.2, label = "Person 2")

plt.xticks(ticks=x_index,labels=xpoints)
plt.ylabel('Day')
plt.xlabel("Walking distance (km)")
plt.title("Persons workout",fontdict = font1)
plt.legend()
```

```python
explode =[0,0,0,0,0.1,0]

plt.pie(ypoints1, labels = xpoints, explode = explode, autopct = '%1.1f%%')
plt.title("Person1 workout by day")
```
